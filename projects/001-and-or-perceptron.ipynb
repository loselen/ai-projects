```python
import numpy as np
import plotly.express as px
```

```python
np.random.seed(67)

w = np.random.randn(2)
b = np.random.randn()

def predict(X):
    global w, b
    z = X @ w + b
    y = np.where(z > 0, 1, 0)
    return y

def train(X, y_true, lr):
    global w, b
    y = predict(X)
    error = y_true - y
    delta_w = lr * (X.T @ error)
    delta_b = lr * np.sum(error)
    w += delta_w
    b += delta_b

X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y = predict(X)

px.scatter(x=X[:, 0], y=X[:, 1], color=y.astype(bool), title="Initial Predictions", width=500, height=500)
```

```python
y_and = np.array([0, 0, 0, 1])

for _ in range(100):
    train(X, y_and, 0.01)

y = predict(X)

px.scatter(x=X[:, 0], y=X[:, 1], color=y.astype(bool), title="AND Predictions", width=500, height=500)
```

```python
y_or = np.array([0, 1, 1, 1])

for _ in range(100):
    train(X, y_or, 0.01)

y = predict(X)

px.scatter(x=X[:, 0], y=X[:, 1], color=y.astype(bool), title="OR Predictions", width=500, height=500)
```
