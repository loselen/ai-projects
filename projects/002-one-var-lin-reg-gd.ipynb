```python
import numpy as np
import plotly.express as px
```

```python
np.random.seed(67)

num_of_students = 30

x = np.random.uniform(0, 7, (num_of_students))
noise = np.random.randn(num_of_students) * 5
y = np.clip(15 * x + noise, 0, 100)

px.scatter(x=x, y=y, width=500, height=500, title="Score vs hours studied data")
```

```python
np.random.seed(67)

w = np.random.randn()
b = np.random.randn()

lr = 0.01

for i in range(100):
    y_pred = x * w + b

    error = y - y_pred

    dMSE_dw = -2 / y.size * (x.T @ error)
    dMSE_db = -2 / y.size * np.sum(error)

    w -= lr * dMSE_dw
    b -= lr * dMSE_db

    if (i + 1) % 10 == 0:
        print(f"Epoch {i + 1}: mse error = {1 / y.size * np.sum(error ** 2)}")

y_pred = x * w + b

fig = px.scatter(x=x, y=y, width=500, height=500, title="Score based on hours studied")
fig.add_scatter(x=x, y=y_pred)
fig
```
